```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Draw a line from position (1,3) and (2,10) then to (6,12) and finally to position (18,20).  
#Mark each point with a beautiful green color and set line colour to red and line style dotted. 

x=[1,2,6,18]
y=[3,10,12,20]

plt.plot(x,y,color='red',linestyle=':',marker='o',markerfacecolor='green')

plt.xlim(0,20)
plt.ylim(0,20)

plt.xlabel("X axis")
plt.ylabel("y axis")

plt.show()
```

```python
#3.Write a python program to draw a line using given axis values taken from a text file, 
#with suitable labels in x and y axis and a title.

file=open("data.csv")
x_data=np.array(list(map(int,file.readline().strip().split(","))))
y_data=np.array(list(map(int,file.readline().strip().split(","))))
file.close()
plt.plot(x_data,y_data,marker='*',color='red',linestyle=':',markerfacecolor='blue')
plt.show()
```

```python
#4. Write a python program to plot two or more lines on same plot with suitable legends for each line. 

x=np.array([1,2,3,4,5])
y_1=np.array([32,33,45,67,80])
y_2=np.array([27,26,29,28,29])
plt.plot(x,y_1,color='blue',marker='.',mfc='red',label='Highest temperature')
plt.plot(x,y_2,color='red',marker='.',mfc='blue',label='lowest temperature')
plt.title("temperature difference")
plt.legend()
plt.show()
```

```python
#5.Write a python program to create multiple plots. 
x1=np.array([1,2,3,4,5])
x2=np.array([3,4,5,6,7])
y1=np.array([6,3,4,5,7])
y2=np.array([8,2,1,3,4])
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("highs")
plt.plot(x1,y1,color='red',marker='x',mfc='pink')
plt.subplot(1,2,2)
plt.title("lows")
plt.plot(x2,y2,color='pink',marker='x',mfc='red')
plt.show()

```

```python
#7.Write a python program to create bar plots of scores by group and gender. Use multiple x values on the 
#same chart for men and women. 
x=np.arange(3,13,2)  #arange(1,6)*3
y1=np.array([22,30,35,35,26])
y2=np.array([25,22,30,35,29])
bar_width=0.4
plt.bar(x-bar_width,y1,label="men",color="blue")
plt.bar(x+bar_width,y2,label="women",color="pink")
plt.legend()
plt.show()
```

```python
#pandas  ---It helps us clean data, handle tables, perform filtering, sorting, grouping, merging, and import/export files like CSV.

#9.

import pandas as pd
df=pd.read_csv("gold.csv")

#stores index col1  col2 call with df['colname']

countries=df['Country'].to_numpy()
medal=df['Gold'].to_numpy()
plt.pie(medal,labels=countries,autopct='%1.1f%%')

# %-starting 1.1f one digit(min) and i decimal part(max) %symbol and % ending
```
