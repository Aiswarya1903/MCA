```python
#9 Given is a dataframe showing name, occupation, salary of people. Find the average  
#salary per occupation
data = [ 
    ['Alice','Teacher',25000], 
    ['Bob','Doctor',55000], 
    ['Charlie','Teacher',20000], 
    ['Dylan','Doctor',35000], 
    ['Emma','Engineer',65000] 
] 

df=pd.DataFrame(data,columns=['name','occupation','salary'])
avg=df.groupby('occupation')['salary'].mean()
print(avg)

#groupby crates group like doctor (5000,3000), [salary] takes the salary from each groups
```

```python
#10.Given a dataframe with NaN Values, fill the NaN values with 0. 
#nan means not a number or missing values
 
 #if orient='index' dict{ p1 []} as it is but if we dont give the orient it will take it as column (the diven each value list as columns)
 #to replace na using fillna()--> so many methods

import numpy as np 
data={
    "p1":["Aishu",100],
    "p2":["Missi",np.nan]
}

df=pd.DataFrame.from_dict(data,orient='index')
df=df.fillna(0)
print(df)
 
 
```

```python
#11.11. Given is a dataframe showing Company Names (cname) and corresponding Profits  
#(profit). Convert the values of Profit column such that values in it greater than 0 are  
#set to True and the rest are set to False. 

data={
    "cname":["A","B","C"],
    "profit":[1000,-500,1]
}

df=pd.DataFrame(data)
df["profit"]=df["profit"]>0
print(df)
```
